```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#test = pd.read_csv('~/ece6254/ECE6254/deepq/save/episode_axis.csv')
#test2 = pd.read_csv('./save/episode_axis.csv')
episodes = pd.read_csv('./save/episode_axis.csv', header=None)
averages = pd.read_csv('./save/rolling_averages.csv', header=None)
epsilons = pd.read_csv('./save/epsilons.csv',header=None)
```

```python
plt.plot(episodes, averages)

plt.show()
```

```python
plt.plot(episodes, epsilons)
plt.show()
```

```python
fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(episodes, averages, 'b-')
ax1.set_xlabel('Episode Number')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Reward', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(episodes, epsilons, 'r.')
ax2.set_ylabel(r'$\epsilon$', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
plt.show()
```
