```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

#test = pd.read_csv('~/ece6254/ECE6254/deepq/save/episode_axis.csv')
#test2 = pd.read_csv('./save/episode_axis.csv')
env = 'Acrobot'
episodes = pd.read_csv('./save/%s/episode_axis.csv' % env, header=None)
averages = pd.read_csv('./save/%s/rolling_averages.csv' % env, header=None)
epsilons = pd.read_csv('./save/%s/epsilons.csv' % env,header=None)

test_episodes = pd.read_csv('./save/%s_test_episode_axis.csv' % env, header=None)
test_averages = pd.read_csv('./save/%s_test_rolling_averages.csv' % env, header=None)
```

```python
test_averages = -test_averages
plt.plot(test_episodes, test_averages, color='C0')
plt.title('Testing %s DeepQ Learning' % env)
plt.xlabel('Episode Number')
plt.ylabel('100 Episode Rolling Average Reward')

plt.show()
```

```python
plt.plot(episodes, epsilons)
plt.show()
```

```python
fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(episodes, averages, 'C0-')
ax1.set_xlabel('Episode Number')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('100 Episode Rolling Average Reward', color='C0')
ax1.tick_params('y', colors='C0')

ax2 = ax1.twinx()
ax2.plot(episodes, epsilons, 'C1.')
ax2.set_ylabel(r'Exploration Rate, $\epsilon$', color='C1')
ax2.tick_params('y', colors='C1')

fig.tight_layout()
plt.title(r'%s, $\gamma = 0.99$' % env)
plt.show()
```
